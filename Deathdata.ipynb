```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data from the CSV file
df = pd.read_csv("Deathdata.csv")
```

```python
# Line Plot
plt.figure(figsize=(20, 15))
plt.subplot(2, 3, 1)
plt.plot(df["Diseases"], df["Deaths"], marker='o')
plt.xticks(rotation=45, ha='right', fontsize=10)  # Adjust rotation, alignment, and font size
plt.title('Deaths by Diseases')
plt.xlabel('Diseases')
plt.ylabel('Deaths')
```

```python
# Bar Chart
plt.figure(figsize=(20, 15))
plt.subplot(2, 3, 2)
plt.bar(df["Diseases"], df["Male"], label='Male')
plt.bar(df["Diseases"], df["Female"], bottom=df["Male"], label='Female')
plt.xticks(rotation=90)
plt.title('Male and Female Deaths by Diseases')
plt.xlabel('Diseases')
plt.ylabel('Deaths')
plt.legend()
```

```python
# Histogram
plt.figure(figsize=(20, 15))
plt.subplot(2, 3, 3)
plt.hist(df["0-4"], bins=10, edgecolor='k')
plt.title('Age Group 0-4 Distribution')
plt.xlabel('Number of Deaths')
plt.ylabel('Frequency')
```

```python
# Box and Whisker Plot
plt.figure(figsize=(20, 15))
plt.subplot(2, 3, 4)
df_age = df.loc[:, "0-4":"70"]
df_age.boxplot(rot=90)
plt.title('Age Group Distribution')
plt.ylabel('Number of Deaths')
```

```python
# Scatter Plot
plt.figure(figsize=(20, 15))
plt.subplot(2, 3, 5)
plt.scatter(df["Deaths"], df["Total"], alpha=0.7)
plt.title('Deaths vs Total Deaths')
plt.xlabel('Deaths')
plt.ylabel('Total Deaths')
```
